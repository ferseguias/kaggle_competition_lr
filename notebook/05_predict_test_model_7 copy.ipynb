```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df = pd.read_csv('../data/train_data_encoded_full_1.csv')
#df.price = np.log(df.price)
print('shape:', df.shape)
df.head(5)
```

```python
X = df.drop(["price"], axis = 1)
y = df['price']

gb = GradientBoostingRegressor(n_estimators=120, loss='squared_error', learning_rate=0.1, max_depth=7, random_state=0, criterion='friedman_mse')
gb.fit(X, y)
```

```python
df_test = pd.read_csv('../data/test_data_encoded_full_1.csv')
df_test
```

```python
#predict test to submit competition
y_pred_gb_test = gb.predict(df_test)
submission_7 = pd.DataFrame(y_pred_gb_test).reset_index()
print('shape:', y_pred_gb_test.shape)
submission_7.columns = ['id', 'price']
submission_7
```

```python
submission_7.to_csv('../output/submission_7_fs.csv', index=False)
```
